##**Task 2: Image to Pencil Sketch with Python**

##**Importing Libraries**

```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

##**Reading the Image**


```python
image = cv2.imread('/content/cup1.jpg') #reading and showing a preview of the image
cv2_imshow(image)
cv2.waitKey(0);
```

##**Converting the Original image to Grayscale Image** 

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #converting into grayscale step
cv2_imshow(gray_image)
cv2.waitKey(0);
```

##**Inverting the new GrayScale Image**

```python
inverted_image = 255 - gray_image #inverting step
cv2_imshow(inverted_image) #basically creating a negative image of the converted grayscale image
cv2.waitKey();
```

##**Blurring the Image by using Gaussian Function in OpenCV**

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21,21),0) #GaussianFunction
```

##**Inverting the Blurred Image in order to see the pencil sketch**

```python
#final step
inverted_blurred = 255 - blurred_image
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale = 256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0);
```
